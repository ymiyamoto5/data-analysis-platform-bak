```python
import csv
import datetime

def main():
    number_of_file = 4
    for i in range(1, number_of_file+1):
        file_name = '3000wave-' + str(i) + '.csv'
        create_outfile(file_name)  
        duplicate_data(file_name, int(200 / number_of_file))
    
def create_outfile(file_name):
  with open(file_name, "w", encoding= "utf-8") as wf:
    writer = csv.writer(wf)
    writer.writerow(["#EndHeader", "日時(μs)", "time(μs）", "(3)HA-V01", "(3)HA-C01"])    

def duplicate_data(file_name, number_of_loop):
  for i in range(0, number_of_loop):
    print(i, datetime.datetime.now())
    with open('wave1-15.csv', "r", encoding = "utf-8") as rf:
      reader = csv.reader(rf)
      _ = next(reader)

      with open(file_name, "a", encoding= "utf-8") as wf:
        writer = csv.writer(wf)
        
        for row in reader:
          writer.writerow(row)

if __name__ == '__main__':
  main()
```
